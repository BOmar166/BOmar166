# Spiel paarungen

```python
def game_setup(player_list):
  players = len(player_list)
  games = int((players / 2) * (players - 1))
  rest = divmod(players, 2)[1]
  rounds = players - 1 + rest
  games_per_round = (players + rest) / 2
  upper_line = []
  lower_line = []
  counter_rounds = 1
  counter_games = 0
  setup = {}

  if rest == 1:
    upper_line.insert(0, "Wildcard")

  for player in player_list:
    upper_line.append(player_list[player][0])
    lower_line.insert(0, player_list[player][0])

  while counter_rounds <= rounds:
    round_list = {}
    #print(f"Runde {counter_rounds}")
    
    while counter_games < games_per_round:
      #print(f"Spiel {(int(games_per_round) - counter_games)}: {upper_line[counter_games]} vs. {lower_line[counter_games]}")
      round_list_temp = round_list
      round_list = {}
      round_list[f"Spiel {(int(games_per_round) - counter_games)}"] = [upper_line[counter_games], lower_line[counter_games]]
      for key in round_list_temp:
        round_list[key] = round_list_temp[key]
      counter_games += 1
    setup[f"Runde {counter_rounds}"] = [round_list]

    upper_line.insert(1, lower_line[0])
    lower_line.pop(0)
    lower_line.append(upper_line[-1])
    upper_line.pop(-1)

    #print("__________________________")
    counter_games = 0
    counter_rounds += 1


  for rounds in setup:
    print(rounds)
    for keys in setup[rounds]:
      for k in keys:
        print(f"{k}: {keys[k][0]} vs. {keys[k][1]}")
    print("__________________________")
  return setup

player_list = {'Player_1': ['Dennis', 0], 'Player_2': ['Micha', 0], 'Player_3': ['Chris', 0], 'Player_4': ['Jasper', 0], 'Player_5': ['Marcel', 0]}
games = game_setup(player_list)
print("")

```

```python
# KO Turnier
player_dict_temp = {'Player_1': ['Micha', 0], 'Player_2': ['Dennis', 0], 'Player_3': ['Marcel', 0], 'Player_4': ['Chris', 0], 'Player_5': ['Jasper', 0]}

import random
answer = ""
ko_player_dict = {}
ko_player_list = []
players = len(player_dict_temp)
rest = divmod(players, 2)[1]
game_rounds = 0

while answer != "e":
  
  print("KO Turnier verwalten:")
  print("[1] Turnierliste erstellen")
  print("[2] Turnierliste anzeigen")
  print("[3] Match eintragen")
  print("[e] Menü verlassen")
  answer = input("Wähle: ")
  print("")
  if answer == "1":

    # Turnierliste erstellen
    if len(player_dict_temp) < 2:
      print("Die Anzahl der Teilnehmer ist zu gering. Mindestens 2 Teilnehmer erforderlich!\n")

    elif ko_player_dict == {}:
      # Wenn die Liste leer ist wird die erste Runde erstellt, ansonsten läuft das Turnier
      # Zufällige Spielerliste erstellen
      ko_player_list = [player for player in player_dict_temp]
      ko_player_list = random.sample(ko_player_list, len(ko_player_list))

      # Wildcard einfügen bei ungrader Spielerzahl
      if rest == 1:
        ko_player_list.insert(0, "Wildcard")
      
      # Spielstruktur erstellen
      game_rounds = 0
      groups_per_round = len(ko_player_list)
      while groups_per_round > 1:
        rest = divmod(groups_per_round, 2)[1]
        if rest == 1:
          groups_per_round += 1
        groups_per_round = int(groups_per_round / 2)

        game_rounds += 1
        Round_list = {}
        groups_per_round_temp = 1
        while groups_per_round_temp <= groups_per_round:    
          Round_list[f"Group_{groups_per_round_temp}"] = []
          groups_per_round_temp +=1

        ko_player_dict[f"Round_{game_rounds}"] = Round_list
        groups_per_round_temp /= 2

      # Gruppen für die erste Runde erstellen
      counter = 0
      Round_1 = {}
      while counter < len(ko_player_list):
        
        print(f"## Gruppe {int(counter / 2) + 1} ##")
        if ko_player_list[counter] == "Wildcard":
          Round_1[f"Group_{int(counter / 2) + 1}"] = {"Wildcard", ko_player_list[counter + 1]}
          print(f"Wildcard vs. {player_dict_temp[ko_player_list[counter + 1]][0]}\n")

        elif ko_player_list[counter + 1] == "Wildcard":
          Round_1[f"Group_{int(counter / 2) + 1}"] = {ko_player_list[counter], "Wildcard"}
          print(f"{player_dict_temp[ko_player_list[counter]][0]} vs. Wildcard\n")

        else:
          Round_1[f"Group_{int(counter / 2) + 1}"] = {ko_player_list[counter], ko_player_list[counter + 1]}
          print(f"{player_dict_temp[ko_player_list[counter]][0]} vs. {player_dict_temp[ko_player_list[counter + 1]][0]}\n")
        
        counter += 2
      ko_player_dict["Round_1"] = Round_1
      
    else:
      print("Das Turnier läuft bereits, erstellen nicht möglich!\n")
   
  elif answer == "2":
    # Turnierliste anzeigen
    if len(ko_player_dict) > 0:
      for round in ko_player_dict:
        round_number = ""
        for number in round:
          if number.isnumeric():
            round_number = round_number + f"{number}"
        print(f"__Runde {round_number}__")

        for group in ko_player_dict[round]:
          player_temp = []
          group_number = ""
          for number in group:
            if number.isnumeric():
              group_number = group_number + f"{number}"
          
          for player in ko_player_dict[round][group]:
            player_temp.append(player)

          if len(player_temp) > 0:
            player_1_temp = player_temp[0]
            player_2_temp = player_temp[1]
            
            if player_1_temp == "Wildcard":
              print(f"Spiel {group_number}: [Wildcard vs. {player_dict_temp[player_2_temp][0]}]\n")
            
            elif player_2_temp == "Wildcard":
              print(f"Spiel {group_number}: [{player_dict_temp[player_1_temp][0]} vs. Wildcard]\n")

            else:
              print(f"Spiel {group_number}: [{player_dict_temp[player_1_temp][0]} vs. {player_dict_temp[player_2_temp][0]}]\n")
           
          else:
            print(f"Spiel {group_number}: [? vs. ?]\n")
    else:
      print("Es wurde noch keine Turnierliste erstellt!\n")

  elif answer == "3":
    # Match eintragen
    print("")

  elif answer == "e":
    # Menü verlassen
    print("KO Turnierverwaltung beendet\n")

  else:
    # Fehleingabe
    print(f"{answer} ist keine gülte Option\n")
```

```python
class Ko_match:
  def __init__(self, player_dict, ko_list, player_1, points_player_1, player_2, points_player_2):

    if player_1 == "Wildcard" or player_2 == "Wildcard":
      self.ko_list = ko_list

    elif points_player_1 > points_player_2:
      #Player 2 wird entfernt
      ko_list.remove(player_2)
      self.ko_list = ko_list
      for player in player_dict:
        if player == player_1:
          player_dict[player][1] +=1
      self.player_dict = player_dict
      
    elif points_player_1 < points_player_2:
      #Player 1 wird entfernt
      ko_list.remove(player_1)
      self.ko_list = ko_list
      for player in player_dict:
        if player == player_2:
          player_dict[player][1] +=1
      self.player_dict = player_dict

player_dict_temp = {'Player_1': ['Micha', 0], 'Player_2': ['Dennis', 0], 'Player_3': ['Marcel', 0], 'Player_4': ['Chris', 0], 'Player_5': ['Jasper', 0]}
ko_player_list = ['Wildcard', 'Player_1', 'Player_3', 'Player_5', 'Player_4', 'Player_2']
match = Ko_match(player_dict_temp, ko_player_list, "Player_3", 2, "Player_4", 1)
print(match.ko_list)
print(match.player_dict)
```
